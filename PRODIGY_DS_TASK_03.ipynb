# Visualization numerical columns using histogram 

```python

df_obj = df.select_dtypes(include = 'object').columns
print(df_obj)
df_num = df.select_dtypes(exclude = 'object').columns
print(df_num)
```

```python
for feature in df_num:
    sns.histplot(x= feature ,data = df, bins = 25 ,kde = True ,color = 'green')
    plt.show()
```

```python
# checking for outliers
df.plot(kind = 'box' , subplots = True ,layout = (5,2) ,figsize = (10,30))
plt.show()
```

```python
columns = ['age','campaign','duration']
for column in columns:
    
        q1 = np.percentile(df[column],25)
        q3 = np.percentile(df[column] ,75)
        iqr = q3-q1
        lower_bound = q1- 1.5*iqr
        upper_bound = q3 + 1.5*iqr
        df = df[(df[column] >=lower_bound) &(df[column]<= upper_bound)]
        

df.plot(kind = 'box',subplots = True , layout = (5,2),figsize = (10,30))
plt.show()
```

```python
high_corr_cols = ['emp.var.rate','euribor3m' ,'nr.employed']
```

```python
# copy the original dataframe
df1 = df.copy()
# Removing high correlated columns from dataset
df1.drop(high_corr_cols,inplace = True ,axis = 1)
df1.columns
```

# showing dimensions of the updated dataset

```python
df1.shape
```

# conversion of the categorical columns into numerical columns  using label encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_encoded =df1.apply(le.fit_transform)
df_encoded
```

```python
if 'subscribed_deposit' in df_encoded.columns:
    print(df_encoded['subscribed_deposit'].value_counts(normalize=True) * 100)
else:
    print("Column 'subscribed_deposit' does not exist in the DataFrame.")
```

```python
# independent variable
x = df_encoded.iloc[:,:-1]
# target variable
y = df_encoded.iloc[:,-1]
```

```python
x.shape
```

```python
y.shape
```

```python
# splitting the dataset into train and test datasets
from sklearn.model_selection import train_test_split
x_train ,x_test,y_train ,y_test  = train_test_split(x,y,test_size = 0.3 ,random_state = 1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Decision Tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dc= DecisionTreeClassifier(criterion = 'gini' ,max_depth = 5 ,min_samples_split = 10)
dc.fit(x_train,y_train)
```

# Evaluating Training and Test Accuracy 

```python
print("Training accuracy:" ,dc.score(x_train,y_train))
print("Testing accuracy: " ,dc.score(x_test ,y_test))
```

```python
y_pred = dc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix , classification_report ,accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test ,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

# Plot Decision Tree

```python
from sklearn.tree import plot_tree
```

```python
feature_names = df.columns.tolist()
plt.figure(figsize  = (40,20))
class_names = ["class_0 ,","class_1"]
plot_tree(dc,feature_names = feature_names ,class_names = class_names ,filled = True ,fontsize = 12)
plt.show()
```
